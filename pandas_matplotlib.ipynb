```python
import math

# Example 1: Calculate the square root
x = math.sqrt(64)
print(f"The square root of 64 is: {x}") # Output: The square root of 64 is: 8.0

# Example 2: Use the constant pi and a function (cos)
print(f"Pi is: {math.pi}") # Output: Pi is: 3.141592653589793
print(f"Cosine of Pi is: {math.cos(math.pi)}") # Output: Cosine of Pi is: -1.0
from math import sqrt, pi, cos

# Example: Use the imported functions directly
print(f"The square root of 25 is: {sqrt(25)}") # Output: The square root of 25 is: 5.0
print(f"Pi is: {pi}") # Output: Pi is: 3.141592653589793
print(f"Cosine of Pi is: {cos(pi)}") # Output: Cosine of Pi is: -1.0
import datetime

# Get current date and time
now = datetime.datetime.now()
print(f"Current Date & Time: {now}")

# Get just the current date
today = datetime.date.today()
print(f"Today's Date: {today}")

# Extract parts of the date
print(f"Year: {today.year}, Month: {today.month}, Day: {today.day}")
```
